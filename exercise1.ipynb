## Exercise 1: Creating and Inspecting the Data

```python
import pandas as pd

# Creating a dictionary of your calendar data
data = {
    'Block': [1, 2, 3, 4, 5, 6],
    'Time_Slot': ['09:00-10:00', '10:15-11:15', '11:30-12:30', '13:30-14:30', '16:00-17:00', '17:15-18:15'],
    'Category': ['Coding', 'Coding', 'Admin', 'Research', 'Coding', 'Admin'],
    'Productivity_Score': [85, 90, 60, 75, 95, 40],
    'Energy_Level': ['High', 'High', 'Medium', 'Medium', 'High', 'Low']
}

# Convert to a DataFrame
df = pd.DataFrame(data)
```
